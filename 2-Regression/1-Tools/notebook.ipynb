# Regression 01 - Tools
### What we do here?
This file works with the built-in [diabetes dataset](https://scikit-learn.org/stable/datasets/toy_dataset.html#diabetes-dataset) from the sklearn and makes a linear regression model plotted as a graph.

### What is linear regression?
*Regression* is a mathematical method used to make predictions about a continuous value. A *linear regression* predicts the value of one variable (called *dependent*) based on the value of another variable (called *independent*). 
It is called linear because the prediction generates a line in the form of a function `y = a + bx`

### The libraries:
- *matplotlib* is a graph plotting tool
- *numpy* handles numeric data on Python
- *sklearn* (*Scikit-learn*) provides tools for supervised and unsupervised learning in ML

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

```python
# Load the diabetes dataset from the sklearn's datasets.
X, y = datasets.load_diabetes(return_X_y=True)
# Printing shape of the matrix (number of items, size of arrays) and the first array 
print(X.shape)
print(X[0])
```

```python
# Selecting the third column of the dataset to plot
X = X[:, 2] # The operator : selects all rows
# It is required a 2D array to plot. To reshape the (rows,cols) of X, we use the reshape function. 
X = X.reshape((-1, 1)) # The parameter -1 calculates the related dimension automatically 
```

```python
# Now that the data is selected, we have to split the independent variable X and the target variable y in two groups: one of test and another of training
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33) # 0.33 means 33% or 1/3 of the entire set
```

```python
# Here comes the training. The fitting function evaluate the model and adequate the necessary parameters with the training set
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
# Create a prediction using the model refined in the previous cell
y_pred = model.predict(X_test)
```

```python
# Plot the prediction. Creates a scatterplot with all the correspondences in X and y test sets
# The line in the middle is an estimate of values for y with a given X values
plt.scatter(X_test, y_test, color="black")
plt.plot(X_test, y_pred, color="blue", linewidth=3)
plt.xlabel("Scaled BMIs")
plt.ylabel("Disease Progression")
plt.title("A Graph Plot Showing Diabetes Progression against BMI")
plt.show()
```
