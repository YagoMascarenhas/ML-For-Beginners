# Regression 02 - Data

### What do we do here?

Here we analyse the [pumpkin dataset](../data/US-pumpkins.csv) in the data folder

### The libraries

- **Pandas** - an open source for data analysis and manipulation

```python
# Importing Pandas and pyplot
import pandas as pd
import matplotlib.pyplot as plt
# Read the csv file
pumpkins = pd.read_csv("../data/US-pumpkins.csv")
# Prints the first five of rows in the dataset
pumpkins.head()
```

```python
# Filter by selecting only the Package with the string "bushel"
pumpkins = pumpkins[pumpkins["Package"].str.contains("bushel", case=True, regex=True)]
pumpkins
```

```python
# Check if there is null (missing) data in the dataset
pumpkins.isnull().sum()
```

```python
# Selecting only the columns we need
# The loc function extracts a part of the dataset. The first parameter is the group of rows you want and the second is a group of columns
columns_to_select = ["Package", "Low Price", "High Price", "Date"]
pumpkins = pumpkins.loc[:, columns_to_select]
pumpkins.head()
```

```python
# Calculates the average price of a punpkin
price = (pumpkins["Low Price"] + pumpkins["High Price"]) / 2
# Convert the Date column to only show the month
month = pd.DatetimeIndex(pumpkins["Date"]).month
print(month)
```

```python
# Copy the converted data to a new dataframe
new_pumpkins = pd.DataFrame({
    "Month": month,
    "Package": pumpkins["Package"],
    "Low Price": pumpkins["Low Price"],
    "High Price": pumpkins["High Price"],
    "Price": price
})
```

```python
# Normalise the prices to show the price for bushel
new_pumpkins.loc[new_pumpkins["Package"].str.contains("1 1/9"), "Price"] = price/(1 + 1/9)
new_pumpkins.loc[new_pumpkins["Package"].str.contains("1/2"), "Price"] = price/(1/2)
new_pumpkins
```

```python
# Displaying the prices as a bar chart grouped by month 
new_pumpkins.groupby(["Month"])["Price"].mean().plot(kind="bar")
plt.ylabel("Pumpkin Price")
```
